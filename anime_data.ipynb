```python
import pandas as pd
import requests
from jikanpy import Jikan
jikan = Jikan()
```

```python
# looking at the response structure.
jikan.anime(14719)
```

```python
fall_anime = jikan.season(year=2020, season='fall')['anime']
```

```python
fall_anime
```

```python
# removing synopsis, since it is not relevant data.
for index, anime in enumerate(fall_anime):
    if 'synopsis' in anime:
        del anime['synopsis']
    
print(fall_anime[5])
```

```python
len(fall_anime) 
```

```python
df = pd.DataFrame(fall_anime)
```

```python
df.head()
```

```python
df.info()
```

```python
# first need to filter away the animes that has had prev seasons ex one peace. 
# Want to be able to recommende new animes that dont have prev seasons.
# and scores with value nan
df = df.dropna(subset=['episodes', 'score'])
```

```python
# dropping movies
df = df[df.type != 'Movie']
```

```python
df_by_score = df.sort_values(['score'], ascending=False)
```

```python
df_score_top10 = df_by_score.head(10)
df_score_bot10 = df_by_score.tail(10)
```

```python
df_clean = pd.concat([df_score_top10, df_score_bot10], ignore_index=True)
```

```python
# for stats"https://api.jikan.moe/v3/anime/123/stats"
# stats save the 'watching' as a new column.
# adding amount of people watching the anime atm

import time

watching = []
for i in range(len(df_clean)):
    r = requests.get('https://api.jikan.moe/v3/anime/{}/stats'.format(df_clean.loc[i]['mal_id']))
    watching.append(r.json()['watching'])
    print(i)
    time.sleep(10)
    
df_clean['watching'] = watching
```
